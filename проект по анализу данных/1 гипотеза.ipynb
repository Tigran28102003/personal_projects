### Гипотеза 1: _в современном баскетболе лучше бросают трехочковые и их влияние на игру стало гораздо больше?_

```python
import numpy as np
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\HUAWEI\hw\проект по анализу данных\данные\df.csv")
```

```python
# создаем копию датасета, чтобы не пришлоось заново парсить данные в случае неудачи, и работаем уже с ней

df_3p = df.copy()

# количество очков, заработанных после трехочковых попаданий
df_3p['point by 3p, pts'] = df_3p['3P'] * 3

# тот же показатель, что и выше только уже в процентах от общего количества очков
df_3p['point by 3p, %'] = (df_3p['3P'] * 3) / df_3p['PTS']
df_3p_agg = df_3p.groupby('year').agg({'point by 3p, pts': ['mean', 'min', 'max', 'std'],
                         'point by 3p, %': ['mean', 'min', 'max', 'std']})
df_3p_agg
```

```python
# создадим отдельный датасет со всеми данными о том, как забивали трехочковые от года к году

tp_stats = df.groupby('year').agg({'3P': ['mean', 'min', 'max', 'std'], 
                        '3PA': ['mean', 'min', 'max', 'std'], 
                        '3P%': ['mean', 'min', 'max', 'std']})

tp_stats = tp_stats.reset_index()


# подгружаем из библиотек все необходимое и визуализируем данне из вышесобранного датасета

import plotly.graph_objs as go
from plotly.subplots import make_subplots

fig = go.Figure()

fig = make_subplots(rows=3, cols=1)

fig.add_trace(go.Scatter(x=tp_stats.year, y=list(tp_stats['3P']['mean']), mode='lines+markers',
                         name='среднее количество забитых по годам'),
               row=1, col=1)
fig.add_trace(go.Scatter(x=tp_stats.year, y=list(tp_stats['3P']['max']), mode='lines+markers',
                    name='максимальное количество забитых по годам'), row=1, col=1)
fig.add_trace(go.Scatter(x=tp_stats.year, y=list(tp_stats['3P']['min']), mode='lines+markers',
                    name='минимальное количество забитых по годам'), row=1, col=1)
fig.add_trace(go.Scatter(x=tp_stats.year, y=list(tp_stats['3P%']['mean'] * 100), mode='lines+markers',
                    name='процент забитых трехочковых по годам'), row=2, col=1)
#fig.add_trace(go.Scatter(x=df_3p.year, y=list(df_3p_agg['point by 3p, pts']['mean']), mode='lines+markers',
 #                   name='среднее количество очков в результатае трехочковых по годам'), row=2, col=1)
#fig.add_trace(go.Scatter(x=df_3p.year, y=list(df_3p_agg['point by 3p, pts']['max']), mode='lines+markers',
 #                   name='максимальное количество очков в результатае трехочковых по годам'), row=2, col=1)
#fig.add_trace(go.Scatter(x=df_3p.year, y=list(df_3p_agg['point by 3p, pts']['min']), mode='lines+markers',
 #                    name='максимальное количество очков в результатае трехочковых по годам'), row=2, col=1)
fig.add_trace(go.Scatter(x=df_3p.year, y=list(df_3p_agg['point by 3p, %']['mean'] * 100), mode='lines+markers',
                    name='средняя доля 3-ёч очковых от среднего кол-ва очков за игру, %'), row=3, col=1)
fig.add_trace(go.Scatter(x=df_3p.year, y=list(df_3p_agg['point by 3p, %']['max'] * 100), mode='lines+markers',
                    name='максимальная доля 3-ёч очковых от среднего кол-ва очков за игру, %'), row=3, col=1)
fig.add_trace(go.Scatter(x=df_3p.year, y=list(df_3p_agg['point by 3p, %']['min'] * 100), mode='lines+markers',
                    name='минимальная доля 3-ёч очковых от среднего кол-ва очков за игру, %'), row=3, col=1)
fig.update_layout(title='Статистика трехочковых по годам')
fig.update_layout(width=1000, height=1200)


fig.update_xaxes(title_text="год", row=1, col=1)
fig.update_xaxes(title_text="год", row=2, col=1)
fig.update_xaxes(title_text="год", row=3, col=1)
fig.update_yaxes(title_text="кол-во забитых 3-ёх очковых", row=1, col=1)
fig.update_yaxes(title_text="процент попадания 3-ёх очковых", row=2, col=1)
fig.update_yaxes(title_text="доля 3-ёч очковых от среднего кол-ва очков за игру, %", row=3, col=1)

fig.show()
```

_легко заметить, что команды стали чаще задействовать атаки с дальных дистанций, хоть изначально и недооценивали это оружие в своем арсенале. хоть в динамике процента попаданий по годам нет систематического роста, в условиях роста среднего количества попыток среди команд, трехочковые стали занимать все больший процент от общего числа очков за игру --> гипотеза о том, что в современном баскетболе несколько неверна, так как процент попаданий остается приблизительно один и тем же, но так как количество попыток увеличилось, мы с точностью можем устверждать, что трехочковые атаки стали неотъемлимой частью игры_
