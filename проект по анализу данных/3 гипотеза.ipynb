```python
import numpy as np
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
import seaborn as sns
import time
from datetime import datetime
from tqdm import tqdm
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_html('https://www.basketball-reference.com/teams/BOS/2023.html')[0]
```

```python
df
```

```python
today = datetime.today().strftime('%Y-%m-%d')
df['Birth Date'] = pd.to_datetime(df['Birth Date'])
cm = []

df['Wt'] = round(df['Wt'] / 2.205, 1)

for hight in df['Ht']:
    hight = hight.split('-')
    ft = int(hight[0])*30.48
    inch = int(hight[1])*2.54
    hight = ft+inch
    cm.append(round(hight, 2))
    
df['Ht'] = cm
```

```python
pos_res = df.groupby('Pos').agg({'Ht': ['mean', 'max', 'min'], 'Wt': ['mean', 'max', 'min']})
pos_res
```
